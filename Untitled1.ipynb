```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path=r'C:\Users\dspra\Downloads\bank.csv'
bank_df=pd.read_csv(file_path,sep=';')

cat=bank_df.select_dtypes(include='object').columns.tolist()
num=bank_df.select_dtypes(exclude='object').columns.tolist()
```

```python

# Take any two from each
two_categorical = cat[:2]
two_numerical = num[:2]

# Create a new DataFrame with selected columns
selected_df = bank_df[two_categorical + two_numerical]

# Display the result
print(selected_df)
```

```python
bank_df['job'].value_counts()
keys=bank_df['job'].value_counts().keys()
values=bank_df['job'].value_counts().values
pd.DataFrame(zip(keys,values),columns=['job','balance'])
```
