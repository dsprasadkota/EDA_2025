```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path=r'C:\Users\dspra\Downloads\visadataset.csv'
visa_df=pd.read_csv(file_path)

cat=visa_df.select_dtypes(include='object').columns
num=visa_df.select_dtypes(exclude='object').columns
```

```python
for i in cat[1:]:
    keys=visa_df[i].value_counts().keys()
    values=visa_df[i].value_counts().values
    plt.figure(figsize=(10,5))
    plt.bar(keys,values)
```

```python
plt.subplot(3,3,1)
plt.subplot(3,3,2)
```

```python
for i in range(1,9):
    plt.subplot(3,3,i)
```

```python
plt.figure(figsize=(30,25))
for i in range(1,9):
    keys=visa_df[cat[i]].value_counts().keys()
    values=visa_df[cat[i]].value_counts().values
    plt.subplot(3,3,i)
    plt.bar(keys,values)
```

```python
plt.figure(figsize=(30,25))
for i in range(1,9):
    keys=visa_df[cat[i]].value_counts().keys()
    values=visa_df[cat[i]].value_counts().values
    plt.subplot(3,3,i).bar(keys,values)
```
